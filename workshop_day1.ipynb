```

import matplotlib.pyplot as plt
```

```
x=[2,9,3,7]
y=[7,15,5,9]
plt.xlabel("distance")  #x axis
plt.ylabel("time")    #y axis
plt.plot(x,y,label='railway track',color='green')  # this will draw a line
plt.scatter(x,y,s=200,label='railway station',marker='*',color='black')  # scatter used for dots and s for size ,marker for marking a point
plt.bar(x,y,label='height of railway station')    # draw a bar chart
plt.grid(color='red')      # for grid line in graph
plt.legend() # show label in the track
plt.show()    #it will draw a graph
```

```
# this will be an another example for cricket data visuligation
player=["dhoni","virat","rohit","sir jad"]
score=[120,140,180,89]

#now ploting it
plt.xlabel('player')
plt.ylabel('score')
#plt.bar(player,score,label="2020 score of india cricket team")
plt.pie(score,labels=player,explode=(0.2,0,0,0),shadow=True,autopct='%1.1ff%%')  # this used for pie chart autopct is a formula for percentage
#plt.grid(color='green')
plt.legend()
plt.show()
```
